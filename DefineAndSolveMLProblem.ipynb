```python
#finding countries with missing values so we can index later. Keeping count as well for own information

countries_and_missing_freedom_counts = {}
countries_with_missing_freedom = []
for index in range(len(unique_countries)):
    if missing_counts[index] > 0:
        countries_and_missing_freedom_counts[unique_countries[index]] = missing_counts[index]
        #countries_with_missing_freedom.append(unique_countries[index])
countries_with_missing_freedom = countries_and_missing_freedom_counts.keys()
print(countries_and_missing_freedom_counts)
print(countries_with_missing_freedom)
```

```python
#replace NaN with average 
df.rename(columns = {'Freedom to make life choices': 'Freedom'}, inplace = True)

#first calculate average of a country with missing data points
for country in countries_with_missing_freedom:
    subset = (df.loc[df['country'] == country])
    subset_mean = subset['Freedom'].mean()
    df.loc[(df['country'] == country) & (df['Freedom'].isna()), 'Freedom'] = subset_mean

(df.loc[df['country'] == 'Vietnam'])
```

```python
#dropping columns with too many missing values 

df = df.drop(columns = ['GINI index (World Bank estimate)', 'gini of household income reported in Gallup, by wp5-year'])
df.columns
```

```python
#let's replace all of the rest of the empty data points with means by country

df = df.rename(columns = {'Log GDP per capita' : 'Log GDP', 'Healthy life expectancy at birth': 'Life expectancy', 'Perceptions of corruption' : 'Corruption'})
empty_columns = ['Log GDP', 'Social support', 'Life expectancy', 'Generosity', 'Corruption', 'Positive affect', 'Negative affect']

df.columns
```

```python
def replace_with_average(column):
    ''' replaces Nan values in a column with the average value for each country'''
    countries_missing = []
    for index in range(len(unique_countries)):
        one_missing_count = df.loc[df['country'] == unique_countries[index]][column].isnull().sum()
        if one_missing_count > 0:
            countries_missing.append(unique_countries[index])
    
    for country in countries_missing:
        subset = (df.loc[df['country'] == country])
        df.loc[(df['country'] == country) & (df[column].isna()), column] = subset[column].mean()
```

```python
for column in empty_columns:
    replace_with_average(column)
```

```python
epochs = [20, 25, 35, 50, 75]
batch_sizes = [5, 10, 15]

results = [] 
metrics = {
    'epochs': [],
    'batch_size': [],
    'train_loss': [],
    'val_loss': [],
    'train_mse': [],
    'val_mse': []
}

for epoch in epochs:
    print("Epochs: " + str(epoch))
    for batch_size in batch_sizes: 
        print("\tBatch Size: " + str(batch_size))
        history = fnn_model.fit(X_train_scaled, y_train, epochs = epoch, batch_size = batch_size, validation_split = 0.2, verbose =0)
        test_loss, test_mse = fnn_model.evaluate(X_test_scaled, y_test)
        #print('Test Loss: ' + str(test_loss))
        #print('Test MSE: ' + str(test_mse))
        val_loss = history.history['val_loss'][-1]
        val_mse = history.history['val_mean_squared_error'][-1]
        results.append((epoch, batch_size, test_loss, test_mse, val_loss, val_mse))

        metrics['epochs'].append(epoch)
        metrics['batch_size'].append(batch_size)
        metrics['train_loss'].append(history.history['loss'][-1])
        metrics['val_loss'].append(val_loss)
        metrics['train_mse'].append(history.history['mean_squared_error'][-1])
        metrics['val_mse'].append(val_mse)
        
# Print the results
print(f"{'Epochs':<10} {'Batch Size':<12} {'Test Loss':<10} {'Test MSE':<10} {'Val Loss':<10} {'Val MSE':<10}")
print('-' * 60)
for epoch, batch_size, test_loss, test_mse, val_loss, val_mse in results:
    print(f"{epoch:<10} {batch_size:<12} {test_loss:<10.4f} {test_mse:<10.4f} {val_loss:<10.4f} {val_mse:<10.4f}")
```
